# Bayesian Engagement View

Estimate the probability that variant B has higher mean engagement than variant A.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

RESULTS_PATH = "data/experiment_results.csv"
METRIC_COLUMNS = ["likes", "replies", "reposts", "bookmarks"]

df = pd.read_csv(RESULTS_PATH)
df["engagement"] = df[METRIC_COLUMNS].sum(axis=1)
df.head()
```

```python
A = df.loc[df["variant"] == "A", "engagement"].to_numpy()
B = df.loc[df["variant"] == "B", "engagement"].to_numpy()

A_mean = A.mean()
B_mean = B.mean()
A_std = A.std(ddof=1)
B_std = B.std(ddof=1)

print("A mean:", round(A_mean, 3))
print("B mean:", round(B_mean, 3))
```

```python
rng = np.random.default_rng(42)
simulations = 50_000

A_samples = rng.normal(A_mean, A_std, simulations)
B_samples = rng.normal(B_mean, B_std, simulations)

prob_B_better = (B_samples > A_samples).mean()
expected_lift = (B_mean - A_mean) / A_mean

print("Probability B is better:", round(prob_B_better, 3))
print("Expected engagement lift:", round(expected_lift, 3))
```

```python
plt.hist(A_samples, alpha=0.5, label="A", bins=30)
plt.hist(B_samples, alpha=0.5, label="B", bins=30)
plt.legend()
plt.title("Simulated Engagement Distributions")
plt.xlabel("Engagement")
plt.ylabel("Simulation count")
plt.show()
```
