# Bayesian-Style Simulation View

Optional simulation view using the same canonical data and config.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from experiment_analysis import load_config, load_results

config = load_config()
df = load_results(config.results_path, config.metric_columns, config.primary_metric)
df.head()
```

```python
control = df.loc[df["variant"] == config.control_variant, config.primary_metric].to_numpy()
treatment = df.loc[df["variant"] == config.treatment_variant, config.primary_metric].to_numpy()

rng = np.random.default_rng(config.random_seed)
simulations = config.bootstrap_iterations
control_samples = rng.normal(control.mean(), control.std(ddof=1), simulations)
treatment_samples = rng.normal(treatment.mean(), treatment.std(ddof=1), simulations)

prob_treatment_better = (treatment_samples > control_samples).mean()
print("Probability treatment is better:", round(prob_treatment_better, 3))
```

```python
plt.hist(control_samples, alpha=0.5, label=config.control_variant, bins=30)
plt.hist(treatment_samples, alpha=0.5, label=config.treatment_variant, bins=30)
plt.legend()
plt.title("Simulated Engagement Distributions")
plt.xlabel(config.primary_metric)
plt.ylabel("Simulation count")
plt.show()
```
