# Power Analysis

Estimate the number of tweets needed per variant for the primary A/B test.


```python
from statsmodels.stats.power import TTestIndPower

baseline_mean = 2
expected_lift = 1.0
std_dev = 2
alpha = 0.05
target_power = 0.8

new_mean = baseline_mean * (1 + expected_lift)
effect_size = (new_mean - baseline_mean) / std_dev

analysis = TTestIndPower()
sample_size = analysis.solve_power(
    effect_size=effect_size,
    power=target_power,
    alpha=alpha,
)

tweets_per_variant = round(sample_size)
total_tweets = tweets_per_variant * 2

print("Tweets per variant:", tweets_per_variant)
print("Total tweets needed:", total_tweets)
```
